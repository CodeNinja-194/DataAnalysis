```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from google.colab import drive
drive.mount('/content/drive',force_remount=True)
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/company_sales_data.csv')
```

```python
display(df)
```

```python
df.info()
monthList  = df ['month_number'].tolist()
faceCremData   = df ['facecream'].tolist()
faceWashData   = df ['facewash'].tolist()
toothPasteData = df ['toothpaste'].tolist()
bathingsoapData   = df ['bathingsoap'].tolist()
shampooData   = df ['shampoo'].tolist()
moisturizerData = df ['moisturizer'].tolist()
total_unitsData=df['total_units'].tolist()                                                                                                                                   
profitList = df ['total_profit'].tolist()
```

```python
#1.	Read Total profit of all months and show it using a line plot.
plt.plot(monthList, profitList, label = 'Month-wise Profit data of last year')
plt.xlabel('Month number')
plt.ylabel('Profit in dollar')
plt.xticks(monthList)
plt.title('Company profit per month')
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.show()
```

```python
#2.	Get total profit of all months and show line plot with the following Style properties
plt.plot(monthList, profitList, label = 'Profit data of last year', color='black',marker='o', linestyle='--')
plt.xlabel('Month Number')
plt.ylabel('Profit in dollar')
plt.legend(loc='lower right')
plt.title('Company Sales data of last year')
plt.xticks(monthList)
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.show()
```

```python
#3.	Read all product sales data and show it using a multiline plot.
plt.plot(monthList, faceCremData,   label = 'Face cream Sales Data', marker='o')
plt.plot(monthList, faceWashData,   label = 'Face Wash Sales Data',  marker='o')
plt.plot(monthList, toothPasteData, label = 'ToothPaste Sales Data', marker='o')
plt.plot(monthList, bathingsoapData, label = 'ToothPaste Sales Data', marker='o')
plt.plot(monthList, shampooData, label = 'ToothPaste Sales Data', marker='o')
plt.plot(monthList, moisturizerData, label = 'ToothPaste Sales Data', marker='o')

plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.xticks(monthList)
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
plt.title('Sales data')
plt.show()
```

```python
#4.	Read toothpaste sales data of each month and show it using a scatter plot
plt.scatter(monthList, toothPasteData,label='Tooth paste Sales data')
plt.xlabel('Month Number')
plt.ylabel('Number of units Sold')
plt.legend(loc='upper left')
plt.title(' Tooth paste Sales data')
plt.xticks(monthList)
plt.grid(True,linestyle="--")
plt.show()
```

```python
# 5.	Read face cream and facewash product sales data and show it using the bar chart
plt.bar(monthList, faceCremData,label ='Face cream Sales Data',width=0.30,align='edge')
plt.bar(monthList, faceWashData,label ='Face Wash Sales Data',width=-0.30,align='edge')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.title(' Sales data')
plt.xticks(monthList)
plt.title('Facewash and facecream sales data')
plt.show()
```

```python
# 6.	Read sales data of bathing soap of all months and show it using a bar chart.
plt.bar(monthList, bathingsoapData,label ='BathingSoap Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Number of units Sold')
plt.legend(loc='upper left')
plt.title(' BathingSoap data')
plt.xticks(monthList)
plt.show()
```

```python
#7.	Read the total profit of each month and show it using the histogram to see the most common profit ranges.
# profitList
labels = ['low', 'average', 'Good', 'Best']
profit_range = [150000, 175000, 200000, 225000, 250000, 300000, 350000]
plt.hist(profitList, profit_range, label = 'Profit data')
plt.xlabel('profit range in dollar')
plt.ylabel('Actual Profit in dollar')
plt.legend(loc='upper left')
plt.xticks(profit_range)
plt.title('Profit data')
plt.show()
```

```python
#8.	Calculate total sale data for last year for each product and show it using a Pie chart
labels = ['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
salesData   = [df ['facecream'].sum(), df ['facewash'].sum(), df ['toothpaste'].sum(), 
         df ['bathingsoap'].sum(), df ['shampoo'].sum(), df ['moisturizer'].sum()]
plt.axis("equal")
plt.pie(salesData, labels=labels)
plt.legend(loc='lower right')
plt.title('Sales data') 
plt.show()
```

```python
#9.	Read Bathing soap facewash of all months and display it using the Subplot.
f, axarr = plt.subplots(2, sharex=True)
axarr[0].plot(monthList, bathingsoapData, label = 'Bathingsoap Sales Data', color='k', marker='o')
axarr[0].set_title('Sales data of  a Bathingsoap')
axarr[1].plot(monthList, faceWashData, label = 'Face Wash Sales Data', color='r', marker='o')
axarr[1].set_title('Sales data of  a facewash')
plt.xticks(monthList)
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.show()
```

```python
# 10.	Read all product sales data and show it using the stack plot
plt.stackplot(monthList, faceCremData, faceWashData, toothPasteData, bathingsoapData, shampooData, moisturizerData, colors=['m','c','r','k','g','y'])

plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.xticks(monthList)
plt.title('Sales data')
plt.show()
```
