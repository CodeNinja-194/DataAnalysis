```python
#5.	List the  cities in which universities were  located.
l=df['Location'].str.split(pat = ',', expand = True)
print(l[0])
```

```python
#6.	List the cities &universities under each state.
df["state"]=l[1]
grouped_by_state = df.groupby("state")
for state, group in grouped_by_state:
    print("State:", state)
    for index, row in group.iterrows():
        print("City:", row["Location"], "University:", row["Location"])
    print("\n")
```

```python
#7.	How many universities in each state have ranking<50? Print them along with ranking.
df_filtered = df.query("Rank<50")
#print(df_filtered)
df["state"]=l[1]
grouped_by_state = df_filtered.groupby("state")
for state, group in grouped_by_state:
    print("State:", state)
    print("Number of universities with ranking less than 50:", group.shape[0])
    for index, row in group.iterrows():
        print("University:", row["Location"], "Ranking:", row["Rank"])
    print("\n")
```

```python
#8.	How many universities have marginal fee difference <=$5000 in in-state &out-of state tuition fees.
print(df.info())
df['Avg'] = df['Tuition and fees'].replace({'\$': '', ',': ''}, regex=True).astype(float)
df['Avg1'] = df['In-state'].replace({'\$': '', ',': ''}, regex=True).astype(float)
#df['Avg'] = df['Tuition and fees'].str.replace({'\$':'',',':''}, regex=True).astype(int)
#df['Avg1'] = df['In-state'].str.replace({'\$':'',',':''}, regex=True).astype(int)
df['diff'] = df['Avg'] - df['Avg1']
df_filtered = df.query('diff <= 5000')
display(df_filtered)
```

```python
#9.	List the universities having tuition fee>15000$ and rank between 120 and  170, inclusive of those  two value
df_filtered = df.query("Avg > 15000 and Rank >= 120 and Rank <= 170")
display(df_filtered)
```

```python
#10.	Find the campuses of universities located in different cities (multiple cities)
l2=df['Name'].str.split(pat = '--', expand = True)
print(l2[1])
```

```python
#11.	Mention the states where out-of state fee is more than in-state fee. Print Minimum and Maximum fees.
df_m= df.query("Avg > Avg1")
display(df_m['Location'])
display(df_m['Tuition and fees'].min())
display(df_m['Tuition and fees'].max())
```

```python
#12.	Find the cities locating top 100 universities.
df3 = df.sort_values("Rank")
top_100 = df3.head(100)
topcities = top_100["Location"].unique()
display("Cities with top 100 universities:", topcities)
```

```python
#13.	Find universities with least no of undergraduate students &highest no of graduate students.
# Find the university with the least number of undergraduate students
min_undergrad_university = df.loc[df["Undergrad Enrollment"] == df["Undergrad Enrollment"].min()]

# Find the university with the highest number of graduate students
max_grad_university = df.loc[df["Undergrad Enrollment"] == df["Undergrad Enrollment"].max()]

# Print out the universities with the least number of undergraduate students and the highest number of graduate students
print("University with least number of undergraduate students:", min_undergrad_university["Name"])
print("University with highest number of graduate students:", max_grad_university["Name"])
```

```python
#14.	Identify correlations between tuition fee and university rank.
correlation = df['Avg'].corr(df['Rank'])
print("Correlation between in_state_tuition and ranking:", correlation)
```
