```python
#1 B)Find out the no. of movies released in every year from 1990 to 1998.
df["Release Date"] = pd.to_datetime(df["Release Date"], errors='coerce')

df1 = df[(df["Release Date"].dt.year >= 1990) & (df["Release Date"].dt.year <= 1998) & (df["Release Date"].notna())]

display(df1)
```

```python
#2 A)Find no. of movies released under each genre given in the database.
grouped = df.groupby("Major Genre").size().reset_index(name="Number of Movies")

print(grouped)
```

```python
#2 B)finding the movies under each genere with IMDB rating >7 and rotten tomato >60

df8 = df[(df["IMDB Rating"] > 7) & (df["Rotten Tomatoes Rating"] > 60)]

# Group the data by the genre column

grouped = df8.groupby("Major Genre").size().reset_index(name="Number of Movies")

print(grouped)
```

```python
#3 A)Find the movies released under each fiction with each director in the ascending order of release dates.
sorted_movies = df.groupby(['Creative Type', 'Director']).apply(lambda x: x.sort_values(by='Release Date', ascending=True))

display(sorted_movies)
#order of the release date 

```

```python
# 3 B)Find movies released under each distributor in the order of genre and director.
#df4=df.groupby("Distributor")
#df4=df.sort_values(by=["Distributor","Major Genre","Director"])
df3=df.groupby('Distributor').apply(lambda x: x.sort_values(by=["Major Genre","Director"]))
display(df3)
```
