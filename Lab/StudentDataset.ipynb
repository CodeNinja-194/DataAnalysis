```python
df1 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/CSE.csv')
df2=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/IT.csv')
df3=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/STUDENT.csv')
display(df1)   
```

```python
#1.	Combine the CSE & IT data and display the data
frames=[df1,df2]
res=pd.concat(frames)
res1= pd.DataFrame(res)
display(res1)
```

```python
#2.	Display all CSE students’ marks along with personal information.
merge_data=df3.merge(df1)
display(merge_data)
```

```python
#3.	Print all students Regd.No, Name and professional elective.
#print(df3.info())
data_concat=df3[['Regd.No','Name']].merge(res1[['Regd.No','Professional Elective']])
display(data_concat) 
```

```python
#4.	Identify the students whose DAA marks are >18.
daa_marks=res1[['Regd.No']].merge(res1[['Regd.No','DAA']])
daa_mark=res1.query('DAA > 18')
print(daa_mark[['Regd.No','DAA']])
```

```python
#5.	Display the names and EAMCET ranks of the students who got minimum 12 marks in all courses.
df2['OE']=12
s=pd.merge(pd.concat([df1,df2]),df3,on='Regd.No')
print(s[(s['DAA']>11)&(s['CN']>11)&(s['AFL']>11)&(s['Professional Elective']>11)&(s['OE']>11)][['Name','EAMCET RANK']])
del df2['OE']
```

```python
#6.	Calculate mean value of all the subject’s marks.
res1.fillna(0,inplace=True)
mean_all=res1[["CN","DAA","AFL","OE","Professional Elective"]].mean()
display(mean_all)
```

```python
#7.	Display the names common in both CSE & IT along with Regd.No.
s=pd.merge(pd.concat([df1,df2]),df3)
k=s.groupby('Name')['Regd.No'].apply(list)
ans={}
for key,value in k.items():
  if len(value)>1: ans[key]=value
print(ans) 

# s=pd.merge(pd.concat([df1,df2]),df3)
# boolean=s['Name'].duplicated()
# print(s[boolean][['Regd.No','Name']])
```

```python
#9.	Divide the students into 5 groups based on average marks.
# c=pd.merge(df1,df2,how='outer')
# c.fillna(0,inplace=True)
# display(c)

s2=pd.concat([df1,df2])
t1=pd.merge(df3,s2)
s2=s2.mean(axis=1,numeric_only=True)
s2=pd.array(s2)
t1['Avg']=s2
d1=pd.cut(s2,5)
t1['Avg_bins']=d1
d=t1[['Regd.No','CN','DAA','AFL','OE','Professional Elective','Avg','Avg_bins']]
display(d)
# pd.merge(s2,t)
# pd.cut(s2,bins=5,labels='bin_values')
```
